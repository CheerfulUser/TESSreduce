```python
import tessreduce as tr
import matplotlib.pyplot as plt
%matplotlib notebook
import numpy as np

plt.rc('text', usetex=True)
plt.rc('font', family='serif')
```

```python
obs = tr.sn_lookup('sn2020cdj',time='max')
tess = tr.tessreduce(obs_list=obs,reduce=True,plot=True,corr_correction=False,savename='sn2020cdj')
```

```python
# subtract the baseline flux
tess.lc[1] -= np.nanmedian(tess.lc[1,:300])
```

```python
fig_width_pt = 240.0  # Get this from LaTeX using \showthe\columnwidth
inches_per_pt = 1.0/72.27               # Convert pt to inches
golden_mean = (np.sqrt(5)-1.0)/2.0         # Aesthetic ratio
fig_width = fig_width_pt*inches_per_pt  # width in inches

plt.rc('text', usetex=True)
plt.rc('font', family='serif')
```

```python
plt.figure(figsize=(3*fig_width,1*fig_width))
plt.subplot(121)

plt.plot(tess.tpf.time.mjd,np.nansum(tess.tpf.flux.value[:,43:48,43:48],axis=(1,2)),'k.',ms=5)
plt.ylabel('Flux ($e^-/s$)',fontsize=15)
plt.xlabel('Time (MJD)',fontsize=15)
plt.xlim(58869.5,58903)

plt.subplot(122)
tess.plotter(time_bin=6/24,ground=True,ax=plt.gca())
plt.xlim(58869.5,58903)
plt.tight_layout()
plt.savefig('full_reduction_example_2020cdj.png')
```
