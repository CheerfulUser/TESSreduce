```python
import tessreduce as tr
import matplotlib.pyplot as plt
%matplotlib notebook
import numpy as np

plt.rc('text', usetex=True)
plt.rc('font', family='serif')

fig_width_pt = 240.0  # Get this from LaTeX using \showthe\columnwidth
inches_per_pt = 1.0/72.27               # Convert pt to inches
golden_mean = (np.sqrt(5)-1.0)/2.0         # Aesthetic ratio
fig_width = fig_width_pt*inches_per_pt  # width in inches
```

```python
tess = tr.tessreduce(tpf='../../../../data/tess/2020fqv.fits',reduce=True)
```

```python
tess.ref_ind
```

```python
plt.figure()#figsize=(1.5*fig_width,3*fig_width))

plt.subplot(131)
plt.title('Reference',fontsize=15)
plt.imshow(tess.ref,vmin=-5,vmax=5,origin='lower')

plt.tick_params(
    axis='both',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    left=False, 
    labelbottom=False,
    labelleft=False) # labels along the bottom edge are off
#ax = plt.gca()


plt.subplot(133)
plt.title('With alignment',fontsize=15)
plt.imshow(tess.flux[tess.ref_ind-40],vmin=-5,vmax=5,origin='lower')

plt.tick_params(
    axis='both',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    left=False, 
    labelbottom=False,
    labelleft=False) # labels along the bottom edge are off
#ax = plt.gca()

plt.subplot(132)
plt.title('Without alignment',fontsize=15)
plt.imshow(tess.tpf.flux[tess.ref_ind+40].value-tess.tpf.flux[tess.ref_ind].value,vmin=-5,vmax=5,origin='lower')

plt.tick_params(
    axis='both',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    left=False, 
    labelbottom=False,
    labelleft=False) # labels along the bottom edge are off
#ax = plt.gca()

plt.tight_layout()
plt.savefig('Alignment_example.pdf', bbox_inches = "tight")
```
